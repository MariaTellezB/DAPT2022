## 3. Functions >> X, y

```python
def manipulate_data(df):

    #2.0 enrollee_id
    enrollee_id = df.enrollee_id
    df.drop('enrollee_id',axis=1,inplace=True)
    
    #2.1 city
    #drop city column, too many categories
    #df.drop(['city'], axis = 1, inplace = True)
    df['city'] = df['city'].str.replace('city_','')
    df.city = pd.to_numeric(df.city)
 
    #2.3 gender
    #"manual imputation":most frequent value
    df['gender'].fillna(value = 'Male', inplace = True)
    df.loc[df.gender == 'Other', 'gender'] = 'Other_g'
    #One Hot Encoder
    gender_encoder = OHpre(sparse = False, handle_unknown = 'ignore')
    gender_encoder_df = pd.DataFrame(gender_encoder.fit_transform(df.gender.values.reshape(-1,1))[:,:], columns = gender_encoder.categories_[0][:])
    #add encoded columns
    df = pd.concat([df, gender_encoder_df], axis = 1)
    #drop original column
    df.drop('gender', axis = 1, inplace = True)
    
    #2.4 relevent_experience
    #No null values >> Label Encoder
    label= LabelEncoder()
    df.relevent_experience = pd.DataFrame(label.fit_transform(df.relevent_experience), columns = ['relevent_experience'])
    
    #2.5 enrolled_university
    #"manual imputation":most frequent value
    df['enrolled_university'].fillna(value = 'no_enrollment', inplace = True)
    #Ordinal Encoder
    ordinal = OrdinalEncoder(categories = [['no_enrollment','Part time course', 'Full time course']], handle_unknown = 'use_encoded_value', unknown_value = np.nan)
    #fit_transform encoded values
    eu = pd.DataFrame (ordinal.fit_transform(df.enrolled_university.values.reshape(-1,1)),columns = ['enrolled_university'])
    #subtitute for encoded values
    df.enrolled_university = eu
    
    #2.6 education_level
    #"manual imputation":most frequent value
    df['education_level'].fillna(value = 'Graduate', inplace = True)
    #Ordinal Encoder
    ordinal_2 = OrdinalEncoder(categories = [['Primary School','High School', 'Graduate','Masters', 'Phd']], handle_unknown = 'use_encoded_value', unknown_value = np.nan)
    #fit_transform encoded values
    el = pd.DataFrame(ordinal_2.fit_transform(df.education_level.values.reshape(-1,1)), columns = ['education_level'])
    #subtitute for encoded values
    df.education_level = el
    
    #2.7 major_discipline
    #"manual imputation":most frequent value
    df['major_discipline'].fillna(value = 'STEM', inplace = True)
    #One Hot Encoder
    major_encoder = OHpre(sparse = False, handle_unknown = 'ignore')
    major_encoder_df = pd.DataFrame(major_encoder.fit_transform(df.major_discipline.values.reshape(-1,1))[:,:], columns = major_encoder.categories_[0][:])
    #add encoded columns
    df = pd.concat([df, major_encoder_df], axis = 1)
    #drop original column
    df.drop('major_discipline', axis = 1, inplace = True)
    
    #2.8 experience
    #subtitute no numerical values
    df.experience = df.experience.fillna(0)
    df.loc[df.experience == '>20', 'experience'] = 20
    df.loc[df.experience == '<1', 'experience'] = 0.5
    #change to float64
    df.experience = pd.to_numeric(df.experience)
    
    #2.9 company_size
    #too many null values, sayonara
    #df.drop(['company_size'], axis = 1, inplace = True)
    #Ordinal Encoder:
    df.company_size = df.company_size.fillna(value = 'no data')
    ordinal_3 = OrdinalEncoder(categories = [['<10','10/49','50-99', '100-500','500-999','1000-4999','5000-9999','10000+','no data']], handle_unknown = 'use_encoded_value', unknown_value = np.nan)
    cs = pd.DataFrame (ordinal_3.fit_transform(df.company_size.values.reshape(-1,1)),columns = ['company_size'])
    df.company_size = cs

    
    #2.10 company_type
    #too many null values, sayonara
   # df.drop(['company_type'], axis = 1, inplace = True)
    df.company_type = df.company_type.fillna(value = 'no data')
    df.loc[df.company_type == 'Other', 'company_type'] = 'Other_ct'
    ct_encoder = OHpre(sparse = False, handle_unknown = 'ignore')
    ct_encoder_df = pd.DataFrame(ct_encoder.fit_transform(df.company_type.values.reshape(-1,1))[:,:], columns = ct_encoder.categories_[0][:])
    df = pd.concat([df, ct_encoder_df], axis = 1)
    df.drop('company_type', axis = 1, inplace = True)
    
    #2.11 
    df.last_new_job = df.last_new_job.fillna(0)
    df.loc[df.last_new_job == '>4', 'last_new_job'] = 5
    df.loc[df.last_new_job == 'never', 'last_new_job'] = 0
    df.last_new_job = pd.to_numeric(df.last_new_job)
    
    
    
    return df
```
